```python
import pandas as pd;
df = pd.read_csv("IRIS.csv")
df
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
sns.histplot(data=df,x='sepal_length')
plt.xlabel("sepal_length")

plt.subplot(2,2,2)
sns.histplot(data=df,x='sepal_width')
plt.xlabel("sepal_width")

# ,hue="species"
plt.subplot(2,2,3)
sns.histplot(data=df,x='petal_length')
plt.xlabel("petal_length")

plt.subplot(2,2,4)
sns.histplot(data=df,x='petal_width')
plt.xlabel("petal_width")
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
sns.boxplot(data=df, x='sepal_length', hue="species")

plt.subplot(2,2,2)
sns.boxplot(data=df, x='sepal_width', hue="species")

plt.subplot(2,2,3)
sns.boxplot(data=df, x='petal_length', hue="species")

plt.subplot(2,2,4)
sns.boxplot(data=df, x='petal_width', hue="species")
```
