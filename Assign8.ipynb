```python
import pandas as pd
df=pd.read_csv('Titanic-Dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
median= df['Age'].median();
df['Age'].fillna(median,inplace=True)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.histplot(data=df,x='Age')
plt.xlabel('Age')
plt.title('Age Distribution')
plt.xticks(range(0, int(df['Age'].max()) + 10, 10))

plt.subplot(1,2,2)
sns.histplot(data=df,x='Fare')
plt.xlabel('Fare')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
sns.histplot(data=df,x='Age',hue='Survived',multiple='stack')
plt.xlabel('Age')
plt.title('Age Distribution')
plt.xticks(range(0, int(df['Age'].max()) + 10, 10))

plt.subplot(2,2,2)
sns.histplot(data=df,x='Fare',hue='Survived',multiple='dodge')
plt.xlabel('Fare')

plt.subplot(2,2,3)
sns.histplot(data=df,x='Age',hue='Sex')
plt.xlabel('Age')

plt.subplot(2,2,4)
sns.histplot(data=df,x='Fare',hue='Sex')
plt.xlabel('Fare')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
sns.boxplot(data=df, x='Survived', y='Age', hue='Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Age Distribution')
plt.xticks(range(2), ['No', 'Yes'])

plt.subplot(2,2,2)
sns.boxplot(data=df,x='Survived',y='Fare',hue='Survived')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.xticks(range(2), ['No', 'Yes'])

plt.subplot(2,2,3)
sns.boxplot(data=df,x='Sex',y='Age',hue='Survived')
plt.xlabel('Sex')
plt.ylabel('Age')

plt.subplot(2,2,4)
sns.boxplot(data=df,x='Sex',y='Fare',hue='Survived')
plt.xlabel('Sex')
plt.ylabel('Fare')
```
