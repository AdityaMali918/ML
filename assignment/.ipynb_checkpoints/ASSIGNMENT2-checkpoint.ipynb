```python
import pandas as pd
import numpy as np

df = pd.read_csv('Book1.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Reading Score'] = df['Reading Score'].fillna(df['Reading Score'].mean())
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
col =['Math Score','Placement Score']

sns.boxplot(data=df,x='Math Score')

plt.subplot(1,2,2)
df.boxplot(col)

plt.show()
```

```python
sns.displot(df['Math Score'])
plt.show()
```

```python
print(df['Math Score'].mean()+3*df['Math Score'].std())
print(df['Math Score'].mean()-3*df['Math Score'].std())
```

```python
df[(df['Math Score']>123) | (df['Math Score']<23)]
```

```python
upper = df['Math Score'].mean()+3*df['Math Score'].std()
lower = df['Math Score'].mean()-3*df['Math Score'].std()

df['Math Score'] = np.where(
    df['Math Score']>upper,
    upper,
    np.where(
    df['Math Score']<lower,
    lower,
    df['Math Score']
)
)

df['Math Score']
```

```python
percentile25 = df['Math Score'].quantile(0.25)
percentile75 = df['Math Score'].quantile(0.75)

iqr = percentile75 - percentile25
iqr

low = (percentile25 - 1.5*iqr)
high = percentile75 + 1.5*iqr

df[(df['Math Score']>high) | (df['Math Score']<low)]
```
